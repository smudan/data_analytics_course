#### Start by importing the necessary libaries and reading the text file into memory. Note: You'll need to use a different separator. Search in google for the documentation on how to do so.

```python
import pandas as pd
df = pd.read_csv('16-17RD.txt', delim_whitespace=True)
df
```

#### What position has the most points? Bonus: Is this statistical significant difference?

```python
df.columns
```

```python
df.groupby('PS')['PTS'].mean()
```

#### What player has the most number of rebounds?

```python
df.sort_values('TR', ascending=False)[['Player', 'TR']]
```

#### What team has the best three point conversion rate? ( made / attempts)

```python
df_grped = df.groupby('Team').mean()[['3M', '3A']]
df_grped['3_conv_rate'] = df_grped['3M'] / df_grped['3A']
df_grped.sort_values('3_conv_rate', ascending=False)
```

```python
df['3_conv_rate'] = df['3M'] / df['3A']
```

```python
df[['Team', '3_conv_rate', '3M', '3A']].sort_values('3_conv_rate', ascending=False)
```

#### What team has the most number of points?

```python
df.groupby('Team').sum()['PTS'].sort_values(ascending=False)
```

#### Create your own custom equation to rank all the team's performance. Include at least three metrics

```python
df.columns
```

```python
df['score_rank'] = df['PTS'] + df['AS'] - df['TO']
df[['Team', 'score_rank']].sort_values('score_rank', ascending=False)
```
