```python
import pandas as pd
import numpy as np
```

```python
nrows = 5000
```

```python
np.random.seed(42)
np.random.binomial(n=1, p=0.5, size=20)
```

```python
def create_impressions_skewed_distribution(size=25):
    pers = range(0, 10)
    # need to parameterize this
    prob=[30.0]*(len(pers)-9) + [5.0]*2 + [1.0]*7
    prob /= np.sum(prob)
    skewed_array = np.random.choice(pers, replace=True, size=size, p=prob)
    return skewed_array
```

```python
def create_clicks_skewed_distribution(size=25):
    pers = range(0, 6)
    # need to parameterize this
    prob=[30.0]*(len(pers)-3) + [5.0]*2 + [1.0]*1
    prob /= np.sum(prob)
    skewed_array = np.random.choice(pers, replace=True, size=size, p=prob)
    return skewed_array
```

```python
def create_day_dist(size):
    # semi-even
    prob = [1.0]*4 + [3.0]*23 + [1.0]*4
    prob /= np.sum(prob)
    return np.random.choice(31, size=size, p=prob)
```

```python
np.random.seed(42)
df = pd.DataFrame({ 'Age' : [int(i) for i in np.random.normal(loc=35, scale=15, size=nrows)],
                    'Gender' : np.random.binomial(n=1, p=0.4, size=nrows),
                    'Impressions' : create_impressions_skewed_distribution(size=nrows),
                    'Clicks' : create_clicks_skewed_distribution(size=nrows),
                    'Signed_In' : np.random.binomial(n=1, p=0.35, size=nrows),
                    'Day' : create_day_dist(size=nrows)})
```
